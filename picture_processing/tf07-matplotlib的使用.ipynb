```python
import numpy as np
import matplotlib.pyplot as plt
x = np.arange(0,10,1)
y = np.array([3,5,7,6,2,6,10,12,15,18])
plt.plot(x,y,'r')  #折线
plt.plot(x,y,'g',lw=5)  #lw  线宽
plt.bar(x,y,1,alpha=1,color='b') #柱状    0.2表示柱宽占比，1时柱子紧挨着
#plt.pie(y)  #饼图，显示同一标签数据占比关系
plt.show()

# alpha 透明度属性只有png图片有
```
